# World Largest Company

**Scraping the data of the world's largest companies by market capitalization**
**from companiesmarketcap.com and exporting the data to a CSV file.**


```python
# Import BeautifulSoup for HTML parsing
from bs4 import BeautifulSoup

# Import requests to send http requests to the website
import requests

# Import pandas for data storage
import pandas as pd
```

```python
# Assign the variable URL the website link
url = "https://companiesmarketcap.com/"

# Send a GET REQUESTS to fetch and get access to the webpage content
webpage = requests.get(url)

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(webpage.text, "html.parser")
```

```python
# Extract the main page title to confirm if the content is correct
header = soup.find("h1", class_="text-center h1-title")

# Print the title text
print(header.text.strip())
```

```python
# Create an empty list to store column names
column_headers = []

# Locate the table header row
column_names = soup.find("tr")

# Extract and store each column header text
for columns in column_names:
    column_headers.append(columns.text)

# Print extracted column headers
print(column_headers)
```

```python
brac =[]
columns_name = soup.find("tr")
for columns in columns_name:
    brac.append(columns.text)
print(brac)
```

```python
# Set pandas to display up to 300 rows for better visibility
pd.set_option("display.max_rows", 50)
```

```python
# Create an empty list to store scraped table rows
rows = []

# Find all table rows in the page
table_rows = soup.find_all("tr")

# Loop through table rows, skipping the header row
for column in table_rows[1:]:
    
    # Extract all table data cells within the row
    table_data = column.find_all("td")
    
    # Clean and extract text from each cell
    row_data = [data.text.strip() for data in table_data]
    
    # Append only rows that match the number of columns
    if len(row_data) == len(brac):
        rows.append(row_data)
        
```

```python
# Convert the scraped data into a pandas DataFrame
DataFrame = pd.DataFrame(rows, columns=brac)

# Displaying the content of the dataframe
DataFrame
```

```python
# Export the DataFrame to a CSV file on the local machine
DataFrame.to_csv("LargeCompaniesByMarketCAP_data.csv", index=False)
```
